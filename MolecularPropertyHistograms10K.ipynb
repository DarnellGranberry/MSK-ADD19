```python
from deepchem.utils.save import load_from_disk
import matplotlib
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
###Display logD histogram
dataset_file= "./enamineSubset10KGroundTruth.csv"
train_dataset_file= "./lipophilicity.csv"

dataset = load_from_disk(dataset_file)
train_dataset = load_from_disk(train_dataset_file)

logDs = np.array(dataset["logD"])
train_logDs = np.array(train_dataset["exp"])

n, bins, patches = plt.hist(logDs, 50, facecolor='green', alpha=.5, label="REAL")
n, bins, patches = plt.hist(train_logDs, 50, facecolor='blue', alpha=0.5, label="TRAIN")

plt.xlabel('Ground truth train and predicted logD')
plt.ylabel('Number of compounds')
plt.title(r'Histogram of train and GT predicted logDs')
plt.grid(True)
plt.legend()
plt.show()
```

```python
###Display ESOL histogram
dataset_file= "./enamineSubset10KGroundTruth.csv"
train_dataset_file= "./esol.csv"

dataset = load_from_disk(dataset_file)
train_dataset = load_from_disk(train_dataset_file)

esol = np.array(dataset["esol"])
train_esol = np.array(train_dataset["measured log solubility in mols per litre"])

n, bins, patches = plt.hist(esol, 50, facecolor='green', alpha=0.5, label="REAL")
n, bins, patches = plt.hist(train_esol, 50, facecolor='blue', alpha=0.5, label="TRAIN")

plt.xlabel('Ground truth predicted and train ESOL')
plt.ylabel('Number of compounds')
plt.title(r'Histogram of predicted and train ESOLs')
plt.grid(True)
plt.legend()
plt.show()
```
